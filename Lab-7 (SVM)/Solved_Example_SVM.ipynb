```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Training data
X = np.array([[1.0, 2.0], [2.0, 1.0]])
y = np.array([1, -1])

# Train the SVM model with a linear kernel
svm = SVC(kernel='linear', C=1.0)
svm.fit(X, y)

# Get the separating hyperplane
w = svm.coef_[0]
b = svm.intercept_[0]

# Calculate slope and intercept of the decision boundary
slope = -w[0] / w[1]
intercept = -b / w[1]

# Plot the decision boundary and margins
plt.figure(figsize=(8, 6))

# Plot the decision boundary
x_values = np.linspace(0, 3, 100)
y_values = slope * x_values + intercept
plt.plot(x_values, y_values, 'k-', label='Decision Boundary ($x_1 - x_2 = 0$)')

# Plot the margins
margin_up = y_values + 1 / w[1]
margin_down = y_values - 1 / w[1]
plt.plot(x_values, margin_up, 'k--', label='Upper Margin ($x_1 - x_2 = +1$)')
plt.plot(x_values, margin_down, 'k--', label='Lower Margin ($x_1 - x_2 = -1$)')

# Plot the support vectors
plt.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], s=200, facecolors='none', edgecolors='k', label='Support Vectors')

# Plot the original data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=100)

# Add labels and title
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title('SVM Decision Boundary with Margins and Support Vectors')

# Annotate the decision boundary equation
plt.text(1.2, 1.2, '$x_1 - x_2 = 0$', fontsize=12, color='black', rotation=45)
# Annotate the upper margin equation
plt.text(1.2, 2.2, '$x_1 - x_2 = +1$', fontsize=12, color='black', rotation=45)
# Annotate the lower margin equation
plt.text(1.2, 0.2, '$x_1 - x_2 = -1$', fontsize=12, color='black', rotation=45)

# Show the legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
