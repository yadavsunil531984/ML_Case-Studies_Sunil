### Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

### Step-1: Load the Dataset

```python
df = pd.read_csv("dataset_1.csv")
print(len(df))
df.head()
```

```python
# Given data and labels
data = np.array([[0, 3], [-1, 0], [1, 2], [2, 1], [3, 3], [0, 0], [-1, -1], [-3, 1], [3, 1]])
labels = np.array([1, 1, 1, 1, 1, -1, -1, -1, -1])
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))

# Plot class +1 points (in blue)
plt.scatter(data[labels == 1, 0], data[labels == 1, 1], color='blue', label='Class +1', s=100)

# Plot class -1 points (in red)
plt.scatter(data[labels == -1, 0], data[labels == -1, 1], color='red', label='Class -1', s=100)

# Set labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Data Points with Labels')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

### Step-2: Train a linear SVM

```python
svm = SVC(kernel='linear', C=1.0)
svm.fit(data, labels)
```

### Step-3: Extract Alpha values for each points

```python
# Retrieve the support vectors and their corresponding alpha values (Lagrange multipliers)
support_vectors_indices = svm.support_
alphas = np.zeros(len(data))  # Initialize alpha values for all points as 0
alphas[support_vectors_indices] = np.abs(svm.dual_coef_[0])  # Assign non-zero alpha values to support vectors

# Print alpha values for all data points
for i, alpha in enumerate(alphas):
    print(f"Data point {data[i]} has alpha value: {alpha}")
```

### Step-4: Extract the weight vector (w) and bias term (b)

```python
w = svm.coef_[0]
b = svm.intercept_[0]

print(f"Weight vector (w): {w}")
print(f"Bias term (b): {b}")
```

```python
print(f"Equation of Decission boundrt are f(x) = {w[0]}.x1 + {w[1]}.x2 + b ")
```

### Step-5: Visualize the data points, decision boundary, and support vectors

```python
# Plot data points with color
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='bwr', s=100)

# Define x1 range for plotting decision boundary
x1_min, x1_max = plt.xlim()

# Generate x1 values for plotting
x1_values = np.linspace(x1_min, x1_max, 100)

# Calculate corresponding x2 values for the decision boundary
x2_values = (-w[0] / w[1]) * x1_values - (b / w[1])

# Plot the decision boundary
plt.plot(x1_values, x2_values, 'k-', label='Decision Boundary')

# Calculate margin lines
margin_up = (-w[0] / w[1]) * x1_values - (b - 1) / w[1]  # Upper margin (positive side)
margin_down = (-w[0] / w[1]) * x1_values - (b + 1) / w[1]  # Lower margin (negative side)

# Plot the margin lines
plt.plot(x1_values, margin_up, 'k--', label='Upper Margin')
plt.plot(x1_values, margin_down, 'k--', label='Lower Margin')

# Fill the margin area between the margin lines
plt.fill_between(x1_values, margin_up, margin_down, color='gray', alpha=0.2)

# Highlight the support vectors
plt.scatter(svm.support_vectors_[:, 0], svm.support_vectors_[:, 1], s=200, facecolors='none', edgecolors='k', label='Support Vectors')

# Display the legend
plt.legend()

# Show the plot
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM with Linear Kernel - Decision Boundary and Margins')
plt.show()
```

### Step-6: Prediction

```python
# Define new points for prediction
new_points = np.array([[1, 1], [0, 2], [-2, 2]])

# Predict the class labels for new points
predictions = svm.predict(new_points)
print(f"Predictions for new points {new_points} are: {predictions}")
```
