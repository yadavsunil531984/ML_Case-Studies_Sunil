### Display data using Scatter Plots

```python
# Given data and labels (assuming we have three classes: 1, 2, and 3)
data = np.array([[0, 3], [-1, 0], [1, 2], [2, 1], [3, 3], [0, 0], [-1, -1], [-3, 1], [3, 1]])
labels = np.array([1, 2, 1, 2, 3, 3, 2, 1, 3])
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))

# Plot class 1 points (in blue)
plt.scatter(data[labels == 1, 0], data[labels == 1, 1], color='blue', label='Class 1', s=100)

# Plot class 2 points (in green)
plt.scatter(data[labels == 2, 0], data[labels == 2, 1], color='green', label='Class 2', s=100)

# Plot class 3 points (in red)
plt.scatter(data[labels == 3, 0], data[labels == 3, 1], color='red', label='Class 3', s=100)

# Set labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Data Points with Three Classes')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

### Train SVM Model

```python
# Train the SVM model (with a linear kernel)
svm = SVC(kernel='linear', decision_function_shape='ovr')  # One-vs-Rest strategy for multi-class classification
svm.fit(data, labels)
```

### Retrieve alpha values

```python
# Retrieve the support vectors and their corresponding alpha values (Lagrange multipliers)
support_vectors_indices = svm.support_
alphas = np.zeros(len(data))  # Initialize alpha values for all points as 0
alphas[support_vectors_indices] = np.abs(svm.dual_coef_[0])  # Assign non-zero alpha values to support vectors

# Print alpha values for all data points
for i, alpha in enumerate(alphas):
    print(f"Data point {data[i]} has alpha value: {alpha}")
```

### Find 'W' and 'b'

```python
w = svm.coef_[0]
b = svm.intercept_[0]

print(f"Weight vector (w): {w}")
print(f"Bias term (b): {b}")
```

### Plot decission boundry

```python
# Plot the decision boundary using mlxtend's plot_decision_regions
plt.figure(figsize=(8, 6))
# Plot decision regions (Note: plot_decision_regions supports only 2D data)
plot_decision_regions(data, labels, clf=svm, legend=2)

# Add plot labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Decision Boundaries for Three Classes')
```

### Prediction

```python
# Predict the class labels for new points
new_points = np.array([[0.5, 2.5], [1, 1], [-2, 0], [3, 2]])
predictions = svm.predict(new_points)

# Print the prediction results
print(f"New Points: {new_points}")
print(f"Predicted Labels for New Points: {predictions}")
```
