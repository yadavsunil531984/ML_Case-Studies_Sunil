### Example-1

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([1.0, 2.0])   #features
y = np.array([300, 500])   #target value
```

```python
linear_model = LinearRegression()
linear_model.fit(x.reshape(-1, 1), y) #X must be a 2-D Matrix
```

```python
b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w}, b = {b:0.2f}")
```

```python
w = w[0]
y_pred = np.dot(w,x) + b
print(f"'manual' prediction: y_pred = wx+b : {y_pred}")
```

### Example-3

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("House_Data.csv")
print(len(df))
df.head()
```

```python
# Convert All Features to Numpy Array
x1 = df['Size (sqft)'].to_numpy()
x2 = df['Number of Bedrooms'].to_numpy()
x3 = df['Number of Floors'].to_numpy()
x4 = df['Age of Home'].to_numpy()
y =  df['Price (1000s dollars)']
x = np.array([x1,x2,x3,x4]).T
```

```python
linear_model = LinearRegression()
linear_model.fit(x, y) 
```

```python
b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w}, b = {b:0.2f}")
```

```python
y_pred = linear_model.predict(x)
print("Prediction on training set:", y_pred)
```

```python
# Optional: Check the first few predictions compared to actual values
comparison_df = pd.DataFrame({'Actual': y, 'Predicted': y_pred})
comparison_df.head()
```

### Example-4 (With Feature Scaled)

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("House_Data.csv")
print(len(df))
df.head()
```

```python
# Convert all features to NumPy arrays
x1 = df['Size (sqft)'].to_numpy()
x2 = df['Number of Bedrooms'].to_numpy()
x3 = df['Number of Floors'].to_numpy()
x4 = df['Age of Home'].to_numpy()
y = df['Price (1000s dollars)'].to_numpy()

# Combine the features into a single NumPy array
x = np.array([x1, x2, x3, x4]).T  # Transpose to shape (n_samples, n_features)

# Apply Z-score normalization (StandardScaler)
scaler = StandardScaler()
z = scaler.fit_transform(x)
```

```python
linear_model = LinearRegression()
linear_model.fit(z, y)
```

```python
b = linear_model.intercept_
w = linear_model.coef_
print(f"w = {w}, b = {b:0.2f}")
```

```python
y_pred = linear_model.predict(z)
print("Prediction on training set:", y_pred)
```

```python
# Optional: Check the first few predictions compared to actual values
comparison_df = pd.DataFrame({'Actual': y, 'Predicted': y_pred})
comparison_df.head()
```
