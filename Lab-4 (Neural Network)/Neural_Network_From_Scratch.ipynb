### Example-3 (Type-2)

```python
import pandas as pd
df = pd.read_csv("cancer_data.csv")
df.head()
```

```python
x1 = df['radius_mean'].to_numpy()
x2 = df['texture_mean'].to_numpy()
y = df['diagnosis'].to_numpy()
x = np.array([x1,x2])
m = len(x1)
```

```python
def dense_Layer1(w1,x,b1):
    a = np.dot(w1,x) + b1
    return a
```

```python
w1 = np.array([[-8.93, -0.1],[0.29,  -7.32],[12.9,  10.81]])
x = np.array([x1,x2])
m = len(x1)
b1 = np.array([[-9.82] * m, [-9.28] * m, [0.96] * m])
a = dense_Layer1(w1,x,b1)
a
```

```python
def output_Layer(w2,a,b2):
    y_pred = np.dot(w2,a) + b2
    return y_pred
```

```python
w2 = np.array([-31.18, -27.59, -32.56])
b2 = 15.41
y_pred = output_Layer(w2,a,b2)
y_pred
```
